```python
from graphviz import Digraph

# Create flowchart
flow = Digraph("YOLOv10n_Pose_Methodology", format="png")
flow.attr(rankdir="TB", size="8")

# Nodes
flow.node("A", "Data Preparation\n- Construction site images/videos")
flow.node("B", "PPE Detection\n- YOLOv10n\n- Compliance vs. Violation")
flow.node("C", "Posture Estimation\n- YOLOv10n-pose\n- Bending, Squatting, Climbing, Normal")
flow.node("D", "Emotion Recognition\n- Face extraction\n- CNN → Alert, Stressed, Distracted")
flow.node("E", "Indoor Localization\n- Homography mapping\n- Restricted & High-risk zones")
flow.node("F", "Explainable AI (XAI)\n- SHAP applied\n- Keypoints & features importance")
flow.node("G", "Output & Reporting\n- Annotated images/videos\n- CSV reports")

# Edges
flow.edges([("A","B"), ("B","C"), ("C","D"), ("D","E"), ("E","F"), ("F","G")])

# Render file
output_path = "/mnt/data/yolov10n_pose_methodology_flowchart"
flow.render(output_path, view=False)

output_path + ".png"
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create figure
fig, ax = plt.subplots(figsize=(10, 12))
ax.axis("off")

# Define flowchart boxes and positions
steps = [
    ("Data Preparation", "Collected construction site images/videos"),
    ("PPE Detection", "YOLOv10n trained for helmet & reflective jacket\nClassified compliance vs. violation"),
    ("Posture Estimation", "YOLOv10n-pose\nDetected 17 human keypoints\nClassified postures: bending, squatting, climbing, normal"),
    ("Emotion Recognition", "Extracted worker faces from detections\nCNN-based classifier → alert, stressed, distracted"),
    ("Indoor Localization", "Homography-based position mapping\nIdentified restricted & high-risk zones"),
    ("Explainable AI (XAI)", "SHAP applied to YOLOv10n-pose outputs\nHighlighted keypoints & features influencing decisions"),
    ("Output & Reporting", "Annotated images/videos\nCSV reports (violations, worker ID, time, location)")
]

# Plot boxes and arrows
y = 1.0
for i, (title, desc) in enumerate(steps):
    # Draw box
    box = mpatches.FancyBboxPatch(
        (0.1, y-0.08), 0.8, 0.1, boxstyle="round,pad=0.02",
        facecolor="#cce5ff", edgecolor="#004085"
    )
    ax.add_patch(box)
    ax.text(0.5, y-0.03, f"{title}\n{desc}", ha="center", va="center", fontsize=9, wrap=True)

    # Draw arrow to next
    if i < len(steps)-1:
        ax.annotate(
            "", xy=(0.5, y-0.1), xytext=(0.5, y-0.2),
            arrowprops=dict(arrowstyle="->", lw=2, color="black")
        )
    y -= 0.2

# Save image
flowchart_path = "/mnt/data/yolov10n_pose_methodology_flowchart_styled.png"
plt.savefig(flowchart_path, bbox_inches="tight")
plt.close()

flowchart_path
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create figure (wide for PPT)
fig, ax = plt.subplots(figsize=(16, 5))
ax.axis("off")

# Define flowchart steps
steps = [
    ("Data Preparation", "Collected construction site images/videos"),
    ("PPE Detection", "YOLOv10n → helmet & reflective jacket\nCompliance vs. Violation"),
    ("Posture Estimation", "YOLOv10n-pose → 17 keypoints\nBending, Squatting, Climbing, Normal"),
    ("Emotion Recognition", "Face extraction\nCNN classifier → Alert, Stressed, Distracted"),
    ("Indoor Localization", "Homography mapping\nRestricted & High-risk zones"),
    ("Explainable AI (XAI)", "SHAP explanations\nKeypoints & features"),
    ("Output & Reporting", "Annotated images/videos\nCSV reports (violations, ID, time, location)")
]

# Draw horizontally
x = 0.02
for i, (title, desc) in enumerate(steps):
    # Draw box
    box = mpatches.FancyBboxPatch(
        (x, 0.4), 0.13, 0.3, boxstyle="round,pad=0.03",
        facecolor="#cce5ff", edgecolor="#004085"
    )
    ax.add_patch(box)
    ax.text(x + 0.065, 0.55, f"{title}\n{desc}", ha="center", va="center", fontsize=9, wrap=True)

    # Draw arrow to next
    if i < len(steps)-1:
        ax.annotate(
            "", xy=(x+0.13, 0.55), xytext=(x+0.15, 0.55),
            arrowprops=dict(arrowstyle="->", lw=2, color="black")
        )
    x += 0.17

# Save horizontal flowchart
flowchart_horizontal = "yolov10n_pose_methodology_flowchart_horizontal.png"
plt.savefig(flowchart_horizontal, bbox_inches="tight")
plt.show()

print(f"Flowchart saved as {flowchart_horizontal}")
```
