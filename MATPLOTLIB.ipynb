```python
price=[12,10,23,12,55000]   #55000 acts as an outliner thus the curve looks like flat for other points and a sudden peak for 55000
year=[2015,2016,2017,2018,2019]
plt.plot(year,price,marker='o',linewidth=3,color='black')
plt.xlabel('years')
plt.ylabel('price')
plt.title('price vs years')
plt.grid()
#this problem is undesirable thus to remove this we limit the scale on the axis
plt.ylim(0,100)
plt.xlim(2015,2020)
```

```python
x = np.linspace(0, 10, 100) #multiple plots= using subplot function

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), 'indigo',linewidth=2)
plt.title("Sine Wave")

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), 'green',linewidth=2)
plt.title("Cosine Wave")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Set random seed for reproducibility
np.random.seed(123)

# Generate sample data (same as R)
x = np.random.normal(loc=50, scale=10, size=100)  # Independent variable
y = 2 * x + np.random.normal(loc=0, scale=5, size=100)  # Dependent variable with noise

# Create scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x=x, y=y, color="blue")
plt.title("Scatter Plot")
plt.xlabel("X Variable")
plt.ylabel("Y Variable")
plt.grid(True)
plt.show()

# Regression model
X = sm.add_constant(x)  # Add constant term for intercept
model = sm.OLS(y, X).fit()
residuals = model.resid  # Get residuals

# Create residual plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x=x, y=residuals, color="red")
plt.axhline(y=0, linestyle="dashed", color="black")  # Zero residual line
plt.title("Residual Plot")
plt.xlabel("X Variable")
plt.ylabel("Residuals")
plt.grid(True)
plt.show()
```
