**  Multiple Linear Regression using our own class**

```python
import numpy as np
import pandas as pd

class MultipleLinearRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        # Convert to numpy arrays
        X = np.array(X)
        y = np.array(y)

        # Number of training examples and features
        m, n = X.shape

        # Initialize weights and bias
        self.weights = np.zeros(n)
        self.bias = 0

        # Gradient Descent
        for _ in range(self.iterations):
            y_pred = np.dot(X, self.weights) + self.bias
            error = y_pred - y

            # Compute gradients
            dw = (1/m) * np.dot(X.T, error)
            db = (1/m) * np.sum(error)

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias


# Load dataset
file_path = "/content/placement.csv"
df = pd.read_csv(file_path).drop(columns=["Unnamed: 0"])

# Define features and target
X = df[["cgpa", "iq"]]
y = df["placement"]

```

# Types of Gradient descent

1.   Batch GD
2.   Stotchastic GD
3.  Mini Batch GD



```python
import numpy as np

def batch_gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)  # Number of training examples
    for _ in range(iterations):
        gradient = (1/m) * X.T @ (X @ theta - y)
        theta -= alpha * gradient
    return theta

# Example Usage
X = np.array([[1, 2], [1, 3], [1, 4]])  # Adding bias term
y = np.array([[3], [6], [9]])
theta = np.zeros((2, 1))
alpha = 0.01
iterations = 1000

theta_final = batch_gradient_descent(X, y, theta, alpha, iterations)
print("Final Theta:", theta_final)
```

```python
def stochastic_gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    for _ in range(iterations):
        for i in range(m):
            xi = X[i:i+1]  # Single training example
            yi = y[i:i+1]
            gradient = xi.T @ (xi @ theta - yi)
            theta -= alpha * gradient
    return theta

theta_final_sgd = stochastic_gradient_descent(X, y, theta, alpha, iterations)
print("Final Theta (SGD):", theta_final_sgd)
```

```python
def mini_batch_gradient_descent(X, y, theta, alpha, iterations, batch_size=2):
    m = len(y)
    for _ in range(iterations):
        indices = np.random.permutation(m)
        X_shuffled, y_shuffled = X[indices], y[indices]
        for i in range(0, m, batch_size):
            X_batch = X_shuffled[i:i+batch_size]
            y_batch = y_shuffled[i:i+batch_size]
            gradient = (1/batch_size) * X_batch.T @ (X_batch @ theta - y_batch)
            theta -= alpha * gradient
    return theta

theta_final_mbgd = mini_batch_gradient_descent(X, y, theta, alpha, iterations, batch_size=2)
print("Final Theta (Mini-Batch GD):", theta_final_mbgd)
```

**Code from scratch**

```python
import numpy as np

class RidgeRegression:
    def __init__(self, alpha=1.0):                                                  #The __init__ method initializes the Ridge Regression model.
        self.alpha = alpha  # Regularization strength
        self.coef_ = None   #initially slope=0
        self.intercept_ = None  #initial intercept=0

    def fit(self, X, y):
        X = np.c_[np.ones(X.shape[0]), X]  # Add bias term                          #Add a column of ones to X to account for the bias term.
        I = np.eye(X.shape[1])                                                      #Create an identity matrix I of the same size as the number of features.
        I[0, 0] = 0  # Don't regularize bias term                                   #Modify I so that the first value is 0. This prevents regularization from affecting the bias term.
        self.weights_ = np.linalg.inv(X.T @ X + self.alpha * I) @ X.T @ y           #Compute the weights using the closed-form solution of Ridge Regression:
        self.intercept_ = self.weights_[0]                                          #Extract the intercept_ (first weight) and coef_ (remaining weights).
        self.coef_ = self.weights_[1:]

    def predict(self, X):
        X = np.c_[np.ones(X.shape[0]), X]  # Add bias term
        return X @ self.weights_

# Example usage
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 3, 5, 7, 11])

model = RidgeRegression(alpha=0.1)
model.fit(X, y)
predictions = model.predict(X)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Predictions:", predictions)
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)  # Insert bias term (1s)
    weights = np.random.randn(X.shape[1])  # Initialize weights randomly
    lr = 0.1  # Learning rate

    for i in range(1000):  # 1000 iterations
        j = np.random.randint(0, X.shape[0])  # Random index
        y_hat = sigmoid(np.dot(X[j], weights))  # Apply sigmoid activation
        weights = weights + lr * (y[j] - y_hat) * X[j]  # Update weights using gradient descent

    return weights[0], weights[1:]  # Return bias and other weights
```
