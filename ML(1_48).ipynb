**Multivariate Imputation by Chained Equations (MICE)**
What is MICE?
Multivariate Imputation by Chained Equations (MICE) is an advanced technique for handling missing values. It iteratively models each feature with missing data using the other features, generating more accurate imputations.

How Does It Work?
Initialize missing values with mean/mode.

Iterate through each column with missing values:

Treat it as the target variable.

Use the other columns as predictors.

Train a regression model to predict missing values.

Replace missing values with predicted values.

Repeat until convergence.

Advantages
✅ Captures relationships between variables.
✅ Works well for datasets with multiple missing values.
✅ More robust than mean/median imputation.

Disadvantages
❌ Computationally intensive.
❌ Can be sensitive to the choice of models and iterations.



```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Sample dataset with missing values
data = {
    'Feature1': [1, 2, np.nan, 4, 5, np.nan, 7, 8, np.nan, 10],
    'Feature2': [5, np.nan, 3, 2, 1, np.nan, 6, 7, np.nan, 8],
    'Feature3': [np.nan, 1, 2, 3, 4, 5, np.nan, 7, 8, np.nan]
}
df = pd.DataFrame(data)
df
```

```python
# Initialize MICE Imputer (IterativeImputer)
mice_imputer = IterativeImputer(max_iter=10, random_state=42)

# Apply imputation
df_imputed = pd.DataFrame(mice_imputer.fit_transform(df), columns=df.columns)

print(df_imputed)
```

# Handling Outliners

**Z-Score**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate a larger dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # Normal distribution (mean=50, std=10)
data = np.append(data, [120, 130, 140])  # Adding outliers

# Create DataFrame
df = pd.DataFrame(data, columns=['values'])
df
```

```python

# Compute Z-scores
df['z_score'] = zscore(df['values'])

# Identify outliers (Z-score > 3 or < -3)
outliers = df[np.abs(df['z_score']) > 3]

# Print results
print("Outliers using Z-Score:\n", outliers)

```

```python
# Define threshold and remove outliers (Z-score > 3 or < -3)
new_df = df[np.abs(df['z_score']) <= 3].drop(columns=['z_score'])
new_df= pd.DataFrame(new_df)
# Print cleaned dataset
print("New DataFrame after Removing Outliers:\n")
new_df
```

**Box Plot and IQR Method**

```python
plt.figure(figsize=(4, 2))
plt.boxplot(df['values'])
plt.title('Box Plot of Values')
plt.ylabel('Values')
plt.grid()
plt.show()
```

```python
# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['values'].quantile(0.25)
Q3 = df['values'].quantile(0.75)

# Compute IQR (Interquartile Range)
IQR = Q3 - Q1
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

import seaborn as sns
sns.distplot(df['values'])
plt.show()
```

```python

# Remove outliers
new_df = df[(df['values'] >= lower_bound) & (df['values'] <= upper_bound)]
new_df= pd.DataFrame(new_df)

# Print cleaned dataset
print("New DataFrame after Removing Outliers:\n")
new_df
```

**Winsorization**


**What is Winsorization?**

Winsorization is a statistical technique used to limit extreme values (outliers) in a dataset by replacing them with the closest non-outlier values instead of removing them. This helps in reducing the impact of extreme outliers while preserving the dataset's overall distribution.

**Algorithm for Winsorization**

Sort the Data: Arrange the dataset in ascending order.

Choose a Threshold (p%): Decide on a percentage (e.g., 5% or 10%) to trim from both ends.

Find the Percentiles: Compute the lower p-th percentile and upper (100-p)-th percentile.

Replace Outliers:

Values below the lower percentile are replaced with the lower percentile value.

Values above the upper percentile are replaced with the upper percentile value.

Return the Modified Dataset.

```python
 import numpy as np
import pandas as pd
from scipy.stats.mstats import winsorize

# Generate sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # Normal distribution (mean=50, std=10)
data = np.append(data, [120, 130, 140])  # Adding outliers

# Create DataFrame
df = pd.DataFrame(data, columns=['values'])

# Apply Winsorization (limits = [lower % cutoff, upper % cutoff])
winsorized_data = winsorize(df['values'], limits=[0.05, 0.05])  # Capping extreme 5% values

# Convert to DataFrame
df['winsorized'] = winsorized_data

# Print Original vs Winsorized Data
print("Original Data:\n", df['values'])
print("\nWinsorized Data:\n", df['winsorized'])

```

# Principal Component Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Generate a random dataset (40 rows, 3 features + 1 target column)
np.random.seed(42) #generates same random numbers everytime
data = np.random.rand(40, 3)*100  # Features scaled between 0-100
target = np.random.randint(0, 2, 40)  # Binary target (0 or 1)

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3'])
df['Target'] = target  # Add target column
df.head(5)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter Plot
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with color based on target
sc = ax.scatter(df['Feature1'], df['Feature2'], df['Feature3'],
                c=df['Target'], cmap='viridis', edgecolor='k')

# Labels
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('3D Scatter Plot of Features')

# Colorbar
plt.colorbar(sc, label='Target Class')

plt.show()
```

```python
# Step 2: Standardize the feature data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Feature1', 'Feature2', 'Feature3']])
df_scaledb=pd.DataFrame(df_scaled)
df_scaledb.head(5)
```

```python
# Step 3: Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components//// 2D
principal_components = pca.fit_transform(df_scaled)
```

```python
# Step 4: Convert PCA result to DataFrame
df_pca = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])
df_pca['Target'] = df['Target']  # Keep target column
df_pca.head(5)
```

```python
# Step 5: Explained variance ratio
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance)
```

```python
# Step 6: Visualizing PCA
plt.figure(figsize=(4,4))
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['Target'], cmap='viridis', edgecolors='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Projection')
plt.colorbar(label='Target Class')
plt.show()
```
